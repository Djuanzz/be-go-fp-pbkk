```python
import pandas as pd
import json
import numpy as np
```

```python
df = pd.read_csv('books.csv')
```

```python
df = df.drop(columns=['subtitle', 'average_rating', 'num_pages', 'ratings_count', 'isbn10'])
```

```python
df = df.dropna(axis=0)
df.info()
```

```python

columns = ["isbn13", "title", "authors", "thumbnail", "description", "published_year"]

df_filtered = df[columns].head(100)

df_filtered = df_filtered.rename(columns={
    "description": "summary", 
    "thumbnail": "book_image", 
    "published_year": "publication_year", 
    "isbn13": "isbn",
    "authors": "author"
})

df_filtered['publication_year'] = df_filtered['publication_year'].astype(int)
df_filtered['isbn'] = df_filtered['isbn'].astype(str)

df_filtered['slug'] = df_filtered['title'].str.lower() \
    .str.replace(' ', '-') \
    .str.replace(r'[^a-z0-9-]', '', regex=True)

df_filtered = df_filtered.sort_values(by='publication_year', ascending=False)

price_values = np.random.choice(np.arange(10000, 300001, 10000), size=len(df_filtered), replace=True)
df_filtered['price'] = price_values.astype(int) 

json_output = df_filtered.to_dict(orient='records')

print(json.dumps(json_output, indent=4))
```
